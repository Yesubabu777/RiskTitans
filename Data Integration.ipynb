```python
import numpy as np
import pandas as pd

def personalized_business_impute(df, value_col, groupings):
    """
    Impute missing values with personalized noise added per group (state, county, incident, etc.).
    The noise factor is calculated based on the group's distribution (mean, std, or range).
    """
    for grouping in groupings:
        # Calculate mean and standard deviation for each group
        group_stats = df.groupby(grouping)[value_col].agg(['mean', 'std', 'count']).reset_index()
        group_stats = group_stats.rename(columns={'mean': 'group_mean', 'std': 'group_std', 'count': 'group_count'})
        
        # Merge the group stats back into the original dataframe
        df = df.merge(group_stats, on=grouping, how='left')
        
        # Impute with group mean
        df[value_col] = df[value_col].fillna(df['group_mean'])
        
        # Add personalized noise: scale noise based on the group's std deviation
        noise_factor = df['group_std'] * 0.2  # You can adjust 0.1 to control the spread
        random_noise = np.random.normal(0, noise_factor, size=df[value_col].shape)
        
        # Apply personalized noise
        df[value_col] = np.maximum(df[value_col] + random_noise,0)
        
        # Drop the temporary columns used for imputation
        df.drop(columns=['group_mean', 'group_std', 'group_count'], inplace=True)

    # Final fallback: fill any remaining nulls with zero or another value
    df[value_col] = df[value_col].fillna(0)  # Adjust this based on business logic
    return df

# Example groupings, adjust as needed for your dataset
groupings = [
    ['fipsStateCode', 'fipsCountyCode', 'incidentType', 'Year'],
    ['fipsStateCode', 'fipsCountyCode', 'incidentType'],
    ['fipsStateCode', 'fipsCountyCode'],
    ['fipsStateCode', 'incidentType'],
    ['fipsStateCode'],
    ['incidentType']
]

# Apply the function to your dataset
df_merged = personalized_business_impute(df_merged, 'DAMAGE_PROPERTY_NUM', groupings)
```

```python
def hierarchical_impute(df, value_col, groupings):
    for grouping in groupings:
        # Compute group average
        group_stats = df.groupby(grouping)[value_col].mean().reset_index()
        group_stats = group_stats.rename(columns={value_col: 'temp_impute'})
        
        # Merge group stats
        df = df.merge(group_stats, on=grouping, how='left')
        
        # Impute only where original is null and new temp value is available
        df[value_col] = df[value_col].fillna(df['temp_impute'])
        df.drop(columns='temp_impute', inplace=True)
        
    # Final fallback: fill any remaining nulls with 0
    df[value_col] = df[value_col].fillna(0)
    return df
```

```python
import pandas as pd
import numpy as np

# Define business types and hazards
business_types = ["Airports & Air Transport", "Hospitals & Healthcare Facilities", 
                  "Power Plants (Electricity Generation & Distribution)", "Water Treatment & Utilities"]
hazards = ["Fire", "Hurricane", "Severe Storm", "Flood", "Tornado", "Snowstorm", 
           "Severe Ice Storm", "Mud/Landslide", "Coastal Storm", "Earthquake"]

# Define time steps (in hours)
time_steps = [0, 12, 24, 36, 48, 60, 72, 84, 96, 108, 120]

# Define states
states = ['Full', 'Partial', 'Non-Operational']

# Define the transition probabilities for each hazard (decay rate)
hazard_decay = {
    "Fire": 0.05, "Hurricane": 0.07, "Severe Storm": 0.06, "Flood": 0.08, 
    "Tornado": 0.1, "Snowstorm": 0.04, "Severe Ice Storm": 0.06, 
    "Mud/Landslide": 0.09, "Coastal Storm": 0.07, "Earthquake": 0.12
}

# Function to simulate the Markov Chain transition
def simulate_markov_chain(initial_state, hazard, time_steps):
    # Initial probabilities based on the starting state
    if initial_state == 'Full':
        full_op = 0.9
        partial_op = 0.1
        non_op = 0.0
    elif initial_state == 'Partial':
        full_op = 0.5
        partial_op = 0.4
        non_op = 0.1
    else:
        full_op = 0.0
        partial_op = 0.2
        non_op = 0.8

    # Transition matrix (simplified)
    transition_matrix = []
    decay_rate = hazard_decay[hazard]

    # Simulate the transitions over time
    simulated_states = [initial_state]
    transition_probs = []

    for t in time_steps[1:]:
        # Transition calculations: reduce Full operation, increase Non-Operational
        decay = decay_rate * (t / 12)
        full_op -= decay
        full_op = max(full_op, 0)
        non_op += decay
        non_op = min(non_op, 1)
        partial_op = 1 - (full_op + non_op)

        # Transition probabilities based on calculated values
        transition_probs.append({
            'Full': full_op,
            'Partial': partial_op,
            'Non-Operational': non_op
        })

        # Determine the next state
        if full_op > partial_op and full_op > non_op:
            simulated_states.append('Full')
        elif partial_op > full_op and partial_op > non_op:
            simulated_states.append('Partial')
        else:
            simulated_states.append('Non-Operational')

    return simulated_states, transition_probs

# Store the results in a list
transition_data = []

# Run simulations for each business and hazard
for business in business_types:
    for hazard in hazards:
        for initial_state in states:
            # Simulate transitions for each initial state over time steps
            simulated_states, transition_probs = simulate_markov_chain(initial_state, hazard, time_steps)
            
            # Store the results with transition probabilities for each state transition
            for time, state, prob in zip(time_steps, simulated_states, transition_probs):
                # For each time step, we need to capture transitions from each state to each other state
                for to_state in states:
                    from_state = initial_state
                    transition_data.append([business, hazard, from_state, to_state, time, prob.get(to_state, 0)])

# Convert to DataFrame
df_markov = pd.DataFrame(transition_data, columns=["Business_category", "incidentType", "From State", "To State", "Time (hrs)", "Transition Probability"])

# Show the final DataFrame
df_markov
```

```python
import pandas as pd
import numpy as np

# Define business types and hazards
business_types = ["Airports & Air Transport", "Hospitals & Healthcare Facilities", "Power Plants (Electricity Generation & Distribution)", "Water Treatment & Utilities"]
hazards = ["Fire", "Hurricane", "Severe Storm", "Flood", "Tornado", "Snowstorm", 
           "Severe Ice Storm", "Mud/Landslide", "Coastal Storm", "Earthquake"]

# Define time steps (in hours)
time_steps = [0, 12, 24, 36, 48, 60, 72, 84, 96, 108, 120]

# Create empty list to store results
data = []

# Function to simulate probabilities based on hazard impact
def generate_probabilities(hazard, time_step):
    """Returns probability values for full, partial, and non-operational states."""
    if time_step == 0:
        return (0.9, 0.1, 0.0)  # Initially, most businesses are fully operational
    
    # Simulating impact: more severe hazards degrade businesses faster
    decay_rate = {
        "Fire": 0.05, "Hurricane": 0.07, "Severe Storm": 0.06, "Flood": 0.08, 
        "Tornado": 0.1, "Snowstorm": 0.04, "Severe Ice Storm": 0.06, 
        "Mud/Landslide": 0.09, "Coastal Storm": 0.07, "Earthquake": 0.12
    }
    
    decay = decay_rate[hazard] * (time_step / 12)  # Increasing effect over time
    
    full_op = max(0.9 - decay, 0.0)  # Decrease full operation probability
    non_op = min(decay, 1.0)  # Increase non-operational probability
    partial_op = 1 - (full_op + non_op)  # Ensure probabilities sum to 1
    
    return (full_op, partial_op, non_op)

# Generate data
for business in business_types:
    for hazard in hazards:
        prev_probs = None  # To compute transition probabilities
        for t in time_steps:
            full_op, partial_op, non_op = generate_probabilities(hazard, t)
            
            if prev_probs:
                # Compute transition probabilities
                full_to_partial = max(prev_probs[0] - full_op, 0)
                full_to_nonop = max(prev_probs[0] - (full_op + full_to_partial), 0)
                partial_to_full = max(prev_probs[1] - partial_op, 0)
                partial_to_nonop = max(prev_probs[1] - (partial_op + partial_to_full), 0)
                nonop_to_partial = max(prev_probs[2] - non_op, 0)
                nonop_to_full = max(prev_probs[2] - (non_op + nonop_to_partial), 0)
                
                # Ensure probabilities are within valid range
                total_transition = full_to_partial + full_to_nonop
                if total_transition > 0:
                    full_to_partial /= total_transition
                    full_to_nonop /= total_transition
                
                total_transition = partial_to_full + partial_to_nonop
                if total_transition > 0:
                    partial_to_full /= total_transition
                    partial_to_nonop /= total_transition
                
                total_transition = nonop_to_partial + nonop_to_full
                if total_transition > 0:
                    nonop_to_partial /= total_transition
                    nonop_to_full /= total_transition
            else:
                full_to_partial, full_to_nonop, partial_to_full, partial_to_nonop, nonop_to_partial, nonop_to_full = [0] * 6
            
            data.append([business, hazard, t, full_op, partial_op, non_op, 
                         full_to_partial, full_to_nonop, partial_to_full, partial_to_nonop, nonop_to_partial, nonop_to_full])
            
            prev_probs = (full_op, partial_op, non_op)  # Store for next transition

# Convert to DataFrame
columns = ["Business_category", "incidentType", "Time (hrs)", "Full Operation Probability", "Partial Operation Probability", 
           "Non-Operational Probability", "P(Full→Partial)", "P(Full→Non-Operational)", "P(Partial→Full)", 
           "P(Partial→Non-Operational)", "P(Non-Operational→Partial)", "P(Non-Operational→Full)"]
df = pd.DataFrame(data, columns=columns)

# Create transition matrix DataFrame
transition_data = []

# Loop to create transition matrix for each combination of business and hazard
for business in business_types:
    for hazard in hazards:
        df_filtered = df[(df['Business_category'] == business) & (df['incidentType'] == hazard)]
        
        # Define state names
        states = ["Full", "Partial", "Non-Operational"]
        
        # Create empty transition matrix
        transition_matrix = np.zeros((3, 3))  # 3 states: Full, Partial, Non-Operational
        
        # Loop through time steps and calculate transitions
        for t in range(1, len(df_filtered)):
            prev_state = np.argmax(df_filtered.iloc[t-1][['Full Operation Probability', 
                                                         'Partial Operation Probability', 
                                                         'Non-Operational Probability']].values)
            curr_state = np.argmax(df_filtered.iloc[t][['Full Operation Probability', 
                                                       'Partial Operation Probability', 
                                                       'Non-Operational Probability']].values)
            
            # Increment the transition matrix
            transition_matrix[prev_state][curr_state] += 1
        
        # Normalize the matrix (rows should sum to 1)
        row_sums = transition_matrix.sum(axis=1, keepdims=True)
        # Avoid division by zero
        row_sums[row_sums == 0] = 1
        transition_matrix = transition_matrix / row_sums
        
        # Add transitions to the data
        for i in range(3):
            for j in range(3):
                transition_data.append([business, hazard, states[i], states[j], transition_matrix[i, j]])

# Convert to DataFrame for transitions
transition_results_df = pd.DataFrame(transition_data, columns=["Business_category", "incidentType", "From State", "To State", "Transition Probability"])

# Display the transition matrix
print(transition_results_df)
```
